### Hipótesis planteadas

```python
#Se utiliza una función para extraer el nombre de la marca de la columna "product_type" creando una columna nueva llamada "marcas"
df['marcas'] = df['product_type'].str.split(' ').str[0]
df['marcas']
```

```python
#Se agrupa por marcas para obtener la sumatoria de ventas diarias de cada una.
ventas_por_marca = df.groupby('marcas')['daily_sold'].sum()
ventas_por_marca.index
```

```python
# Grafico en barras las marcas y sus cantidades vendidas

# Crear el gráfico de torta usando la variable de ventas diarias por marca y su índice
fig, ax = plt.subplots(figsize=(10, 10))
ax.pie(ventas_por_marca, labels=ventas_por_marca.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
ax.set_title('Cantidad de celulares vendidos segun su marca')

plt.show()
```

Tal como se muestra en el gráfico anterior, las marcas más vendidas no fueron las propuestas en la hipótesis, sino que aparecieron Samsung y Xiaomi como las más requeridas por los usuarios.

### Diferencia de ventas con pago contra entrega vs pago previo al despacho:

```python
# Filtrar los datos por forma de pago contra entrega (Verdadero o Falso)
df_pagoCE = df[df['cod'] == True]
df_pagoPREV = df[df['cod'] == False]

# Cantidad de ventas realizadas con pago contra entrega
VentasCE = df_pagoCE['total_sold'].sum()
Ventas_sinPCE = df_pagoPREV['total_sold'].sum()

# Crear los datos para los gráficos de torta
ventas_totales = [VentasCE, Ventas_sinPCE]
labels = ['Pago Contra Entrega', 'Pago Previo al Despacho']

# Crear el gráfico de torta
fig, ax = plt.subplots(figsize=(7, 7))
ax.pie(ventas_totales, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
ax.set_title('Distribución de Ventas: Pago Contra Entrega vs Pago Previo al Despacho')

plt.show()
```

Como se puede observar, hay una clara diferencia entre los clientes que pagan al momento de la entrega del producto respecto de aquellos que pagan antes de recibirlo. Estos ultimos representan la mayoria de los casos.

```python
# Filtrar ventas diarias de las marcas más vendidas
df_Samsung = df[df['product_type'].str.contains('Samsung', regex = False)]
df_Xiaomi = df[df['product_type'].str.contains('Xiaomi', regex = False)]


# Subplot con dos boxplots dentro en forma paralela
fig, axes = plt.subplots(2, 1, figsize=(12, 12))

# Boxplot para Samsung 
axes[0].boxplot(df_Samsung['daily_view'], vert=False)
axes[0].set_title('Ventas diarias Samsung')
axes[0].set_xlabel('Samsung')

# Boxplot para Xiaomi
axes[1].boxplot(df_Xiaomi['daily_view'], vert=False)
axes[1].set_title('Ventas diarias Xiaomi')
axes[1].set_xlabel('Xiaomi')

plt.show()
```

En el grafico anterior no puede observarse el bigote dado que los datos de la media como la desviacion estándar (una al 9 y otra al 18) aparecen entre el 0 y el 20. Con una muestra de, por ejemplo 26000 casos, el bigote deja  de verse. Esto significa que todas las ventas, si bien pueden llegar a ser 40000 en un día en su mayoría dan entre 0 y 100 aprox. Vale también aclarar que la dispersión de ventas diarias para Samsung es bastante menor a la correspondiente para Xiami, debido a la menor concentración de puntos observados entre las 2500 y 5000 ventas.

### Lugares y Ventas 



```python
# Agrupación de lugares con mayor venta

df_lugar = df.groupby('merchant_city').size()
df_ventas = df.groupby('daily_sold')


# Agrupar por ciudad y sumar las ventas diarias
df_ventas_lugar = df.groupby('merchant_city')['daily_sold'].sum().reset_index()

# Ordenar los valores para mejor visualización
df_ventas_lugar = df_ventas_lugar.sort_values(by='daily_sold', ascending=False)

# Crear el gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(df_ventas_lugar['merchant_city'], df_ventas_lugar['daily_sold'], color='blue')
plt.xticks(rotation=90)
plt.title('Ventas diarias por ciudad')
plt.xlabel('Ciudad')
plt.ylabel('Total de ventas diarias')
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Gráfico de Seaborn para observar correlaciones


```python
sns.pairplot(df, diag_kind="kde")
```

Los gráficos anteriores demuestran cierto grado de correlación entre total_sold que es la totalidad de ventas respecto de total_review que refiere a las respuestas de los usuarios sobre los productos. También se observa una relación directa entre (daily_sold) que son los dias en los que se vendieron los productos y el margen de ganancia producido por los mismos (daily_gmv). 

```python
# Matriz de correlación de variables
def plotCorrelationMatrix(df, graphWidth):
    dataset_name = 'Marketing_celulares'
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    
    plt.title(f'Matriz de correlaciones para {dataset_name}', fontsize=15)
    plt.show()
```

```python
plotCorrelationMatrix(df,10)
```

De acuerdo a la matriz de correlaciones que se observa anteriormente, encontramos una correlación entre total_review y total_sold, lo cual nos permite advertir que las respuestas o feedback de los usuarios influyen en las ventas de los productos. Por otro lado, también encontramos que hay una relación entre las variables: daily_sold, daily view y daily gmv, las cuales remiten a ventas por dia, margen de ganancia por dia y visitas a la paginas web.
